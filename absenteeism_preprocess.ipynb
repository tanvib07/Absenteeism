For the reasons we can form groups wrt the type of reasons ,in order to categorize it.<br>
The groups are 1-14,15-17,18-21,22-28

```python
r1 = reason_col.iloc[:,0:15].max(axis=1)
r2 = reason_col.iloc[:,15:18].max(axis=1)
r3 = reason_col.iloc[:,18:22].max(axis=1)
r4 = reason_col.iloc[:,22:29].max(axis=1)
```

```python
data = pd.concat([data,r1,r2,r3,r4], axis=1)
```

```python
data
```

```python
data.columns.values
```

```python
col_names=['Reason for Absence', 'Date', 'Transportation Expense',
       'Distance to Work', 'Age', 'Daily Work Load Average',
       'Body Mass Index', 'Education', 'Children', 'Pets',
       'Absenteeism Time in Hours', 'Reason1', 'Reason2', 'Reason3', 'Reason4']
data.columns = col_names
```

```python
col_names = [ 'Reason1', 'Reason2', 'Reason3', 'Reason4','Reason for Absence', 'Date', 'Transportation Expense',
       'Distance to Work', 'Age', 'Daily Work Load Average',
       'Body Mass Index', 'Education', 'Children', 'Pets',
       'Absenteeism Time in Hours']

data = data[col_names]
data.head()
```

```python
data["Date"] = pd.to_datetime(data["Date"], format = '%d/%m/%Y')
```

```python
data["Date"]
```

```python
month_list = []

for i in range(len(data["Date"])):
    month_list.append(data["Date"][i].month)

month_list
```

```python
data["Month value"] = month_list
data.head()
```

```python
week_list = []

for i in range(len(data["Date"])):
    week_list.append(data["Date"][i].weekday())

week_list
```

```python
data["Day of week"] = week_list
```
