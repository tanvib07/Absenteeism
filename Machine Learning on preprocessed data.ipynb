```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("preprocess.csv")
```

```python
data.head()
```

```python
targets = np.where(data["Absenteeism Time in Hours"]>data["Absenteeism Time in Hours"].median(),1,0)
targets
```

```python
data["Excessive Absenteeism"] = targets
data.head()
```

```python
data = data.drop(["Absenteeism Time in Hours"], axis=1)
data= data.drop(["Date"], axis=1)
data= data.drop(["Daily Work Load Average","Distance to Work","Day of week"], axis=1)
```

```python
data_final = data.copy()
data_final.head()
```

```python
x = data_final.iloc[:,:-1]
y = data_final.iloc[:,-1]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaled_x = scaler.fit(x)
scaled_x = scaler.transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(scaled_x,y,test_size=0.2, random_state=42)
```

### Logistic regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
reg = LogisticRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```

#### Finding intercept and coeff

```python
reg.intercept_
```

```python
reg.coef_
```

```python
feature_names = x.columns.values
```

```python
summary_table = pd.DataFrame(columns=['Feature_names'],data=feature_names)
summary_table['Coefficient'] = np.transpose(reg.coef_)  #transpose as nd arrays are rows not columns
summary_table
```

```python
summary_table.index = summary_table.index +1
summary_table.loc[0] = ['Intercept',reg.intercept_[0]]
summary_table = summary_table.sort_index()
summary_table
```

```python
summary_table['Odds_ratio'] = np.exp(summary_table.Coefficient)
summary_table
```

```python
summary_table.sort_values('Odds_ratio',ascending=False)
```

#### Prediction of accuracy

```python
reg.score(x_test,y_test)
```
